```python
dummy = pd.get_dummies(raw_data[['Make', 'Model', 'Trim', 'Type']])
dummy #Model_9-2X AWD, Make_Saturn, Type_Convertible
```

```python
fdata = pd.concat([raw_data, dummy], axis = 1)
```

```python
fdata.drop(labels = ['Make', 'Model', 'Trim', 'Type'], axis = 1, inplace = True)
```

```python
fdata.shape
```

```python
fdata.info()
```

```python
df1 = fdata.var().sort_values(ascending = False).reset_index()
df1.to_csv('variance_01.csv')
```

```python
df1.columns
```

```python
df1.rename(columns=({'index' : 'name', 0 : 'values'}), inplace = True)
```

```python
fdata.drop(df1.query("values < 0.05")['name'].to_list(), axis = 1, inplace = True)
```

```python
fdata.shape
```

```python
fdata.corr().to_csv('Corr01.csv')
```

```python
fdata.drop(['Liter', 'Type_Coupe', 'Type_Sedan'], axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

import statsmodels.api as sm
```

```python
df_train, df_test = train_test_split(fdata, test_size = 0.3, random_state = 2)
```

```python
df_test.shape, df_train.shape
```

```python
x_train = df_train.drop('Price', axis = 1)
```

```python
y_train = df_train['Price']
```

```python
x_train.shape, y_train.shape
```

```python
X_train = sm.add_constant(x_train)
lm1 = sm.OLS(y_train, X_train).fit()
lm1.summary()
```

```python
X_train01 = x_train.drop(['Cruise', 'Type_Hatchback', 'Leather'], axis = 1)
```

```python
X_train02 = sm.add_constant(X_train01)
lm2 = sm.OLS(y_train, X_train02).fit()
lm2.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
X = X_train02.drop('const', axis = 1)
vif['features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
X_train02.drop(['Make_Chevrolet', 'const'], axis = 1, inplace = True)
```

```python
X_train03 = sm.add_constant(X_train02)
lm3 = sm.OLS(y_train, X_train03).fit()
lm3.summary()
```

```python
y_train_predict = lm3.predict(X_train03)
```

```python
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
train_result = y_train - y_train_predict
plt.axvline(x=0, color = 'blue')
plt.xlabel('Training Errors', fontsize = 14)
sb.distplot(train_result, rug = True, color = 'red')
```

```python
r2_score(y_train, y_train_predict)
```

```python
train_result_sq = train_result * train_result
train_result_sqrt = np.sqrt(train_result_sq)
```

```python
np.mean(train_result_sq), np.mean(train_result_sqrt)
```

```python
np.mean(train_result_sqrt)/np.mean(y_train) * 100
```

```python
X_train03.columns
```

```python
cols = ['Mileage', 'Cylinder', 'Doors', 'Sound', 'Make_Buick',
       'Make_Cadillac', 'Make_Pontiac', 'Make_SAAB', 'Make_Saturn',
       'Model_AVEO', 'Model_Cavalier', 'Model_Cobalt', 'Model_Ion',
       'Model_Malibu', 'Trim_Coupe 2D', 'Trim_LS Sedan 4D', 'Trim_Sedan 4D',
       'Type_Convertible', 'Type_Wagon']
f_test = df_test[cols]
y_test_org = df_test['Price']
```

```python
final_test_vars = sm.add_constant(f_test)
```

```python
y_test_pred = lm3.predict(final_test_vars)
```

```python
test_result = y_test_org - y_test_pred
plt.axvline(x = 0)
plt.xlabel('Test Errors', fontsize = 14)
sb.distplot(test_result, rug = True, color = 'orange')
```

```python
(r2_score(y_test_org, y_test_pred) * 100)
```

```python
test_result_sq = test_result * test_result
test_result_sqrt = np.sqrt(test_result_sq)
```

```python
np.mean(test_result_sqrt) / np.mean(y_test_org) * 100
```

```python
plt.xlabel('Y Test', fontsize = 14)
plt.ylabel('Y Test Predicted', fontsize = 14)
plt.scatter(y_test_org, y_test_pred, color = 'red')
```

```python
predicted_data = pd.concat([y_test_org, y_test_pred], axis = 1)
predicted_data.to_csv('Predicted_Data01.csv')
```
